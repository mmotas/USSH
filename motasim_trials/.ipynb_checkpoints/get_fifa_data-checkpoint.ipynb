```python
import pandas as pd
```

```python
def get_player_ratings():
    PlayerRatings = pd.read_csv("../../hackathon data/reduced data/2018.csv", encoding = "ISO-8859-1")
    pr = {}
    for player, rating in zip(PlayerRatings['Name'], PlayerRatings['Overall Rating']):
        pr[player] = rating
    return pr
```

```python
def get_attributes():
    fp = open("attributes.txt")
    attributes = fp.realines()
    attributes = [attribute.replace("\n","") for attribute in attributes]
    return attributes
```

```python
PlayerRatings = get_player_ratings()
```

```python
data = pd.read_csv("../../hackathon data/Hackathon Raw Files/Full Datasets - Opta/USMNT/USMNT.csv", encoding = "ISO-8859-1")
matches = {}
data = data.fillna(0)
matchids = data['game_id']
playerids = data['player_id']
playernames = data['player']
for mid, pid, pn in zip(matchids,playerids,playernames):
    print(matches)
    if mid not in matches.keys():
        matches[mid] = {}
        matches[mid][pid] = PlayerRatings[pn]
    else:
        matches[mid][pid] = PlayerRatings[pn]
```
