```python
def get_player_ratings():
    PlayerRatings = pd.read_csv("../../hackathon data/reduced data/use.csv", encoding = "ISO-8859-1")
    pr = {}
    for player, rating in zip(PlayerRatings['Name'], PlayerRatings['Overall Rating']):
        player = player.lower()
        print(player)
        pr[player] = rating
    pr[0] = "UNK"
    pr["0"] = "UNK"
    return pr
```

```python
def get_attributes():
    fp = open("attributes.txt")
    attributes = fp.realines()
    attributes = [attribute.replace("\n","") for attribute in attributes]
    return attributes
```
